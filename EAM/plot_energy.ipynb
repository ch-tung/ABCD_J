```python
import matplotlib.pyplot as plt
import numpy as np
def plot_energy_vs_steps(filename):
    # Read the lines from the file
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Strip the newline character and convert to float
    energies = [float(line.strip()) for line in lines]

    # Create a list of steps
    steps = list(range(1, len(energies) + 1))

    # Plot the energy vs steps
    plt.plot(steps, energies)
    plt.xlabel('Steps')
    plt.ylabel('Potential energy (eV)')
    plt.title('Potential energy vs steps')
    plt.show()
```

```python
plot_energy_vs_steps('output.txt')
```

```python
def plot_energy_vs_steps_p(filename):
    # Read the lines from the file
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Strip the newline character and split each line by space
    data = [line.strip().split() for line in lines]

    # Convert the data to float
    energies = np.array([float(data[0][0])]+[float(line[0]) for line in data])
    energies_p = np.array([0]+[float(line[1]) for line in data])
    
    # Create a list of steps
    steps = list(range(0, len(energies)))

    # Plot each line separately
    plt.plot(steps, energies, '-k', label='Potential energy')
    plt.plot(steps, energies+energies_p, '-r', label='Potential energy with penalty')

    plt.xlabel('Steps')
    plt.ylabel('Potential energy (eV)')
    plt.title('Potential energy vs steps')
    plt.legend(frameon=False)
    plt.show()
```

```python
plot_energy_vs_steps_p('output_constrain.txt')
```
